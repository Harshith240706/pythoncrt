```python
names=["sai","kalyan","laxmi","naru","karthik"]
marks=[[30,40,50],[66,77,55],[33,78,90],[87,97,57],[47,88,44]]
for i in range(5):
  p=sum(marks[i])//3
  if p >= 80:
    grade="outstanding(A)"
  elif p >= 60:
    grade="average(B)"
  elif p >= 50:
    grade="Poor(c)"
  else:
    grade="fail"
  print("{}.Mr/Ms {} has scored {}% - {} grade".format(i+1,names[i].title(),p,grade))
```

```python
import random

a=int(input("Enter num of teams:"))
teams=[]
print("Enter all teams name:")
for i in range(a):
  t=input("Enter team name:")
  teams.append(t)
m=int(input("Enter num of meet bw teams: "))

matches=[]
for i in range(0,a-1):
  for j in range(i+1,a):
    for k in range(m):
      matches.append([teams[i],teams[j]])

random.shuffle(matches) # Shuffle once after all matches are generated

pos=1
for i in matches:
  print("MATCH {} : {} vs {}".format(pos,i[0],i[1])) # Corrected print statement
  pos=pos+1
```
