```python
legs=int(input("Enter num of legs:"))
heads=int(input("Enter num of heads:"))
flag=False
for cows in range(0,heads+1):
  hens=heads-cows
  cal_legs=cows*4+hens*2
  if cal_legs==legs:
    print("cows:",cows)
    print("hens:",hens)
    flag=True
    break
  if flag==False:
      print("NO solution")
```
